FROM Orders

This outputs a table where it gives the total number of times each employee ordered something and that is counted in the TotalRep column.

The answer would be employee with ID # 4.

What product was ordered the most by customers in Germany?

SELECT p."ProductID", p."ProductName", SUM(o."quantity") AS quantity
FROM Customers, OrderDetails AS o
    INNER JOIN "Products" AS p
    ON o."ProductID" = p."ProductID"
WHERE Country = "Germany"
GROUP BY o."ProductID"
ORDER BY SUM(o."quantity") DESC, p."ProductName" ASC

In this case, the product that sold the most in Germany would be ProductID 31, Gorgonzola Telino at 5038 Units.
